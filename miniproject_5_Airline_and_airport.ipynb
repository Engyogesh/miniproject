```python
data = pd.read_excel(r'/content/gdrive/My Drive/mini project/airports_analysis.xlsx')
display(pd.DataFrame(data))
df = pd.DataFrame(data)
```

```python
1."Top 20 highest altitude airports in the world. Choose only “type” with “airport”."

df_airports = df[['airport_id', 'name', 'city', 'country', 'altitude']]
df_airports = df_airports.sort_values(by='altitude', ascending=False).head(20)
df_airports['rank'] = range(1, 21)

df_airports_result = df_airports[['airport_id', 'name', 'city', 'country', 'altitude','rank']]

print(tabulate(df_airports_result, headers='keys', tablefmt='grid'))

```

```python
2."Top 20 countries that have the highest total # of airports."

df_country_airports = df.groupby('country').size().reset_index(name='Total_Airports')
df_country_airports = df_country_airports.sort_values(by='Total_Airports', ascending=False).head(20)
df_country_airports['Rank'] = range(1, 21)

print(tabulate(df_country_airports, headers='keys', tablefmt='grid'))

```

```python
3."Top 20 cities that have the highest total # of airports. (Two cities from different countries can have the same name, and you have to make sure the numbers are not messed up due to that!)"

df_city_airports = df.groupby(['city', 'country']).size().reset_index(name='Total_Airports')
df_city_airports = df_city_airports.sort_values(by='Total_Airports', ascending=False).head(20)
df_city_airports['rank'] = range(1, 21)

print(tabulate(df_city_airports, headers='keys', tablefmt='grid'))
```

```python
data = pd.read_excel(r'/content/gdrive/My Drive/mini project/airline.xlsx')
display(pd.DataFrame(data))
df_flights = pd.DataFrame(data)
```

```python
1."In the year 2009, print Top 10 cities in the USA that received maximum # inbound flights."

top_inbound_2009 = df_flights[['city', 'total_inbound_flights']].sort_values(by='total_inbound_flights', ascending=False).head(10)
top_inbound_2009['inbound_rank'] = range(1, 11)

print("Top 10 Cities by Inbound Flights in 2009:")
print(tabulate(top_inbound_2009, headers='keys', tablefmt='grid'))
```

```python
2."In the year 2009, print Top 10 cities in the USA that received maximum # outbound flights."

top_outbound_2009 = df_flights[['city', 'total_outbound_flights']].sort_values(by='total_outbound_flights', ascending=False).head(10)
top_outbound_2009['outbound_rank'] = range(1, 11)

print("\nTop 10 Cities by Outbound Flights in 2009:")
print(tabulate(top_outbound_2009, headers='keys', tablefmt='grid'))
```

```python
1."Read the input csv file from here or here and convert columns to rows.
The "Value" column values should come from the crop column names such as Maize Production (tonnes), Rice Production (tonnes) from the input csv file.
(hint: UnPivot melt)."

melted_data = pd.melt(data, id_vars=['Entity', 'Year'], var_name='Crop', value_name='Value')

print(tabulate(melted_data.head(), headers='keys', tablefmt='grid'))
```

```python
2."For each unique value in column "Productions", you have to assign a unique integer/float value. The following is an example output with unique float value. You are not supposed to use custom function. You are not supposed to provide explicit mapping between values in "Produtions" and integer/float values"

data['Productions'] = (
    data['Maize Production (tonnes)'].astype(str) + " " +
    data['Rice  Production ( tonnes)'].astype(str) + " " +
    data['Yams  Production (tonnes)'].astype(str) + " " +
    data['Wheat Production (tonnes)'].astype(str)
)

# Factorize the unique values in the 'Productions' column
data['Productions_Factor'], unique_productions = pd.factorize(data['Productions'])

# Convert the factorized integers to floats
data['Productions_Float'] = data['Productions_Factor'].astype(float) / len(unique_productions)
print(data[['Productions', 'Productions_Float']])
```
